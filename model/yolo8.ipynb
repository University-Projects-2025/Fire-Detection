## YOLOv8

I tried the train the YOLOv8n with 10 epoch on CPU, but it was about 6 hours and later i used GPU accelerator

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(
    data='../data/dataset/dataset.yaml',
    epochs=10,
    imgsz=640,
    batch=16
)
```

## Validation of YOLOv8

```python
from ultralytics import YOLO

model = YOLO('runs/detect/train_yolo8n/weights/best.pt')
metrics = model.val()

print(metrics)
```
